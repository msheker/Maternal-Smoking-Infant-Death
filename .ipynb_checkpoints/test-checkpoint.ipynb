```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open("babies.-1ibz3w3-2h9hx4o.txt")

header = 0
lines = []

for l in f:
    if header == 0:
        header = l.split(" ")
        lines.append([x.strip('\n') for x in header])
    else:
        ln = []
        for d in l.split(" "):
            if( d == '' or d == "\n"):
                continue
            ln.append(int(d.strip('\n')))
        lines.append(ln)
f.close()

nf = open("babies.csv", "w")

for vector in lines:
    strl = ""
    for elem in vector:
        strl = strl + str(elem) + ","
    strl = strl[:-1] + "\n"
    nf.write(strl)

nf.close()
```

```python
data = pd.read_csv("babies.csv")

#note 2 outliers exist
data = data[data["age"] < 90]
#note 13 outliers exist
data = data[data["gestation"] < 500]

data.sort_values(['bwt'])

data_s = data[data["smoke"] == 1]
data_ns = data[data["smoke"] == 0]
```

```python
data.plot(x="age", y="bwt", kind="scatter")
data.plot(y="bwt", x="gestation", kind="scatter")
data.plot(x="age", y="gestation", kind="scatter")
plt.show()
```

```python
data_s.plot(y="bwt", x="gestation", kind="scatter", grid=True, xticks=range(120,400,40),yticks = range(40,200,20))
data_ns.plot(y="bwt", x="gestation", kind="scatter", grid=True, xticks=range(120,400,40),yticks = range(40,200,20))

plt.show()
```
